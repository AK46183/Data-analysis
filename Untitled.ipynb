```python
import pandas as pd

df = pd.read_excel("Coffee Shop Sales.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.shape)
```

```python
df.columns
```

```python
df["Revenue"] = df["transaction_qty"] * df["unit_price"]
```

```python
df.head()
```

```python
total_revenue = df["Revenue"].sum()
total_transactions = df["transaction_id"].nunique()
average_sale = df["Revenue"].mean()

print("Total Revenue:", total_revenue)
print("Total Transactions:", total_transactions)
print("Average Sale:", average_sale)
```

```python
top_products = (
    df.groupby("product_detail")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

```python
import os

os.makedirs("charts", exist_ok=True)
os.makedirs("reports", exist_ok=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

top_products.plot(kind="bar")

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig("charts/top10_products.png")

plt.show()
```

```python
category_sales = df.groupby("product_category")["Revenue"].sum()

plt.figure(figsize=(8,5))

category_sales.plot(kind="bar")

plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig("charts/category_revenue.png")

plt.show()
```

```python
summary = {
    "Metric": [
        "Total Revenue",
        "Total Transactions",
        "Average Sale"
    ],
    "Value": [
        total_revenue,
        total_transactions,
        average_sale
    ]
}

report = pd.DataFrame(summary)

report.to_excel("reports/Automation_Report.xlsx", index=False)
```
