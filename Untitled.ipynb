### Chi-Square Test

```python
from scipy.stats import chi2_contingency

# Contingency Table
contingency_table = pd.crosstab(df["test group"], df["converted"])

print(contingency_table)

# Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
```

```python
if p < 0.05:
    print("Reject H0: There is a significant difference between the campaigns.")
else:
    print("Fail to Reject H0: There is no significant difference between the campaigns.")
```

### T-Test

```python
from scipy.stats import ttest_ind

# Separate groups
ad_group = df[df["test group"] == "ad"]["converted"]
psa_group = df[df["test group"] == "psa"]["converted"]

# Independent T-Test
t_stat, p_value = ttest_ind(ad_group, psa_group)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
    print("There is a significant difference between Campaign A (ad) and Campaign B (psa).")
else:
    print("Fail to Reject H0")
    print("There is no significant difference between the campaigns.")
```

#### The statistical analysis shows that the Ad Campaign performed better than the PSA Campaign. Since the p-value is less than 0.05, the difference is statistically significant. Therefore, the null hypothesis is rejected, and the Ad Campaign is recommended because it generates a higher conversion rate.
